```python
import pandas as pd

# If your file is CSV:
df = pd.read_csv('/content/Student_feedback.csv')

# OR if your file was Excel:
# df = pd.read_excel('/content/student_feedback.xlsx')

# Now check
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace('\n', '', regex=True)
```

```python
df.columns
```

```python
df.rename(columns={
    'How would you rate your overall experience at the School Festival?': 'Experience',
    'Which part of the festival did you enjoy the most?': 'EnjoyedMost',
    'Did you find the festival well organized?': 'Organized',
    'Was the timing and schedule convenient for you?': 'TimingConvenience',
    'How was the cleanliness and overall environment during the event?': 'Cleanliness',
    'What did you like the most about the School Festival?': 'Likes',
    'What improvements would you suggest for next year’s festival?': 'Improvements',
    'Was there any event or activity you think should be added next time?': 'NewEvent',
    'Would you like to attend the School Festival again next year?': 'AttendNextYear',
    'Any other comments or suggestions you’d like to share?': 'Comments'
}, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np  # for tick control

sns.countplot(x='Experience', data=df, palette='viridis')

plt.title('Overall Experience of Students', fontsize=14)
plt.xlabel('Experience Rating (1–5)', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)

# 🔹 Make Y-axis show only natural numbers (1, 2, 3, …)
y_max = df['Experience'].value_counts().max()
plt.yticks(np.arange(0, y_max + 1, 1))

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np  # for tick control

sns.countplot(x='Cleanliness', data=df, palette='viridis')

plt.title('Overall Cleanliness of Students', fontsize=14)
plt.xlabel('Cleanlines Rating (1–5)', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)

# 🔹 Make Y-axis show only natural numbers (1, 2, 3, …)
y_max = df['Cleanliness'].value_counts().max()
plt.yticks(np.arange(0, y_max + 1, 1))

plt.show()
```

```python
df.groupby('CLASS')['Cleanliness'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Cleanliness by Class')
plt.ylabel('Average Rating')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(
    y='EnjoyedMost',
    data=df,
    order=df['EnjoyedMost'].value_counts().index,
    palette='coolwarm'
)

plt.title('Most Enjoyed Parts of the Festival', fontsize=14)
plt.xlabel('Number of Students', fontsize=12)
plt.ylabel('Festival Part', fontsize=12)
plt.show()
```

```python
# Create a function to get sentiment polarity
def get_sentiment(text):
    if pd.isnull(text):
        return None
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0.1:
        return 'Positive'
    elif analysis.sentiment.polarity < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

# Apply on Likes and Improvements columns
df['Likes_Sentiment'] = df['Likes'].apply(get_sentiment)
df['Improvements_Sentiment'] = df['Improvements'].apply(get_sentiment)

# Plot Likes sentiment
sns.countplot(x='Likes_Sentiment', data=df, palette='pastel')
plt.title('Sentiment of Likes Comments')
plt.show()

# Plot Improvements sentiment
sns.countplot(x='Improvements_Sentiment', data=df, palette='pastel')
plt.title('Sentiment of Improvement Suggestions')
plt.show()
```

```python
# Define a function to get polarity score (-1 = negative, +1 = positive)
def get_sentiment(text):
    if pd.isna(text):
        return 0
    return TextBlob(text).sentiment.polarity

# Apply sentiment analysis to both columns
df['Likes_Sentiment'] = df['Likes'].apply(get_sentiment)
df['Improvements_Sentiment'] = df['Improvements'].apply(get_sentiment)

# Display average sentiments
print("Average sentiment for 'Likes':", df['Likes_Sentiment'].mean())
print("Average sentiment for 'Improvements':", df['Improvements_Sentiment'].mean())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
sns.histplot(df['Likes_Sentiment'], bins=10, kde=True, color='green', label='Likes')
sns.histplot(df['Improvements_Sentiment'], bins=10, kde=True, color='red', label='Improvements')
plt.title('Sentiment Distribution of Student Feedback', fontsize=14)
plt.xlabel('Sentiment Polarity (-1 = Negative, +1 = Positive)')
plt.ylabel('Number of Comments')
plt.legend()
plt.show()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Replace missing values with an empty string
df['Likes'] = df['Likes'].fillna('')

# Make text lowercase and remove unnecessary symbols
df['likes_clean'] = df['Likes'].str.lower().str.replace('[^\w\s]', '', regex=True)
```

```python
text = " ".join(df['likes_clean'])
```

```python
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    colormap='viridis',     # color theme
    max_words=100
).generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Positive Feedback Words', fontsize=16)
plt.show()
```

```python
df['Improvements'] = df['Improvements'].fillna('')
df['improvements_clean'] = df['Improvements'].str.lower().str.replace('[^\w\s]', '', regex=True)

text_improve = " ".join(df['improvements_clean'])
wordcloud_improve = WordCloud(width=800, height=400, background_color='white', colormap='Reds').generate(text_improve)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud_improve, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Improvement Suggestions', fontsize=16)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
from wordcloud import WordCloud
```

```python
plt.figure(figsize=(15,10))

# Plot 1 – Satisfaction / Experience rating
plt.subplot(2,2,1)
sns.countplot(x='Experience', data=df, palette='viridis')
plt.title('Overall Experience Rating')
plt.xlabel('Rating (1–5)')
plt.ylabel('Number of Students')

# Plot 2 – Sentiment comparison (Likes vs Improvements)
plt.subplot(2,2,2)
sns.kdeplot(df['Likes_Sentiment'], fill=True, color='green', label='Likes')
sns.kdeplot(df['Improvements_Sentiment'], fill=True, color='red', label='Improvements')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Polarity (-1 to +1)')
plt.legend()

# Plot 3 – Word Cloud: Likes
plt.subplot(2,2,3)
text_likes = " ".join(df['likes_clean'])
wc_likes = WordCloud(width=600, height=400, background_color='white', colormap='Greens').generate(text_likes)
plt.imshow(wc_likes, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Positive Feedback Words')

# Plot 4 – Word Cloud: Improvements
plt.subplot(2,2,4)
text_improve = " ".join(df['improvements_clean'])
wc_improve = WordCloud(width=600, height=400, background_color='white', colormap='Reds').generate(text_improve)
plt.imshow(wc_improve, interpolation='bilinear')
plt.axis('off')
plt.title('Most Common Improvement Suggestions')

plt.tight_layout()
plt.show()
```

```python
print("📊 FINAL INSIGHTS SUMMARY\n")

print(f"✅ Average Satisfaction Rating: {df['Experience'].mean():.2f} / 5")
print(f"💚 Average Sentiment (Likes): {df['Likes_Sentiment'].mean():.2f}")
print(f"❤️ Average Sentiment (Improvements): {df['Improvements_Sentiment'].mean():.2f}")

positive_comments = (df['Likes_Sentiment'] > 0).sum()
negative_comments = (df['Improvements_Sentiment'] < 0).sum()

print(f"\n👍 Positive Feedback Comments: {positive_comments}")
print(f"👎 Negative or Suggestive Comments: {negative_comments}")

print("\n🎯 Suggested Focus for Next Event:")
print("- Keep promoting the most loved parts (top words in Likes WordCloud).")
print("- Improve the areas mentioned often in the Improvements WordCloud.")
print("- Maintain good organization and cleanliness based on overall satisfaction ratings.")
```
